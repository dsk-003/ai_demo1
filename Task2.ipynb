
# **Exploratory Data Analysis On  the EV Cars dataset**

## The goal of this assignment is to analyse and visualise the Electric Vehicle dataset using Python programming.(Matplotlib, seaborn and plotly)


## **About Dataset**
This dataset shows the Electric Vehicles:
Data Fields:

- Battery:   The capacity of the vehicle's battery in kilowatt-hours (kWh).

- Car_ name: The model’s name of the electric vehicle.

- Car_ name_ link: A direct link to the corresponding page on the EV Database for more in-depth information.

- Efficiency: The energy efficiency rating of the vehicle in watt-hours per kilometer (Wh/km).

- Fast_ charge: The fast-charging capability of the vehicle in minutes for a certain charging percentage.

- Price DE. : The price of the electric vehicle in Germany.

- Range: The driving range of the vehicle on a single charge in kilometers.

- Top_ speed: The maximum speed the vehicle can achieve in kilometers per hour.

- Acceleration..0.100.: The acceleration time from 0 to 100 kilometers per hour.

```python
# 1. Filtering based on a condition (e.g., cars with a range over 500)

filtered_cars = data_df[data_df['Range'] > 500]
print("\nCars with a range over 500:")
print(filtered_cars[['Car_name', 'Range']])
```

### **Step 3: Handling missing values**





```python
data_df.isna().sum()
```

```python
data_df.nunique()
```

```python
data_df.isnull().sum()
```

```python
# Fill null values in 'Price.DE.' column with the mean value

mean_price = data_df['Price.DE.'].mean()
data_df['Price.DE.'].fillna(mean_price, inplace=True)
```

```python
# Fill null values in 'Fast_charge' column with the mean value

mean_fast_charge = data_df['Fast_charge'].mean()
data_df['Fast_charge'].fillna(mean_fast_charge, inplace=True)

```

### **Step 4: Data Visualization**

```python
# plot heatmap of int and float values to check data correlation

data_df.columns
```

```python
# Heatmap is a graphical way to represent data. It is most commonly used in data analysis.

df2=data_df[['Battery','Efficiency', 'Fast_charge','Price.DE.', 'Range', 'Top_speed', 'acceleration..0.100.']]
plt.figure(figsize=(15,8))
sns.heatmap(df2.corr(),annot=True,cmap='coolwarm')
```

```python
# plot the countof brand appears

plt.figure(figsize=(15,8))
sns.countplot(x='Top_speed',data=data_df,palette='coolwarm')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.title('Speed wise vehicle count')
plt.show()
```

```python
data_df.columns
```

```python
#display the price of top 10 expensive cars

plt.figure(figsize=(15,8))
sns.barplot(x='Car_name',y='Price.DE.',data=data_df.sort_values(by='Price.DE.',ascending=False).head(10),palette='rainbow')
plt.xticks(rotation=90)
plt.title('Top 10 expensive cars')
plt.grid(axis='y')
plt.show()
```

```python
#display the price of least 10 expensive cars

plt.figure(figsize=(15,8))
sns.barplot(x='Car_name',y='Price.DE.',data=data_df.sort_values(by='Price.DE.',ascending=True).head(10),palette='rainbow')
plt.xticks(rotation=90)
plt.title('Least 10 expensive cars')
plt.grid(axis='y')
plt.show()
```

```python
# Visual Bar chart - Top 10 cars by Range

top_10_range = data_df.nlargest(10, 'Range')
plt.figure(figsize=(10, 6))
plt.barh(top_10_range['Car_name'], top_10_range['Range'], color='skyblue')
plt.xlabel('Range')
plt.title('Top 10 Cars by Range')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Visual Scatter plot - Range vs Price

plt.figure(figsize=(8, 6))
plt.scatter(data_df['Range'], data_df['Price.DE.'], color='salmon', alpha=0.7)
plt.xlabel('Range')
plt.ylabel('Price (in DE)')
plt.title('Range vs Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### **Step 5: Insights Obtained**


#### As per the above graph we can analyse the most of battery powered cars are of price 50k to 70k with distance covering range of 300 to 500.

```python
#The histogram gives an understanding of how battery capacities are distributed across the dataset.
#It shows the concentration or spread of different battery capacities and their frequencies within the provided data.

# Visualization of  Histogram - Distribution of Battery Capacity

plt.figure(figsize=(8, 6))
plt.hist(data_df['Battery'], bins=15, color='lightgreen', edgecolor='black')
plt.xlabel('Battery Capacity')
plt.ylabel('Frequency')
plt.title('Distribution of Battery Capacity')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#The boxplot allows for a comparison of the distribution of Battery Capacities across different Top Speed categories.
# Boxplot of Battery Capacity by Top Speed

plt.figure(figsize=(12, 6))
sns.boxplot(x='Top_speed', y='Battery', data=data_df, palette='pastel')
plt.title('Boxplot of Battery Capacity by Top Speed')
plt.xlabel('Top Speed (km/h)')
plt.ylabel('Battery Capacity (kWh)')
plt.show()
```

```python
# Pie chart representation of the distribution of Top Speed categories within the dataset.
#It helps in quickly grasping the relative proportions of different Top Speed levels in the dataset.

# Pie chart of the distribution of Top Speed categories

plt.figure(figsize=(8, 8))
top_speed_counts = data_df['Top_speed'].value_counts()
plt.pie(top_speed_counts, labels=top_speed_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Top Speed Categories')
plt.show()
```

```python
# Analyzing this histogram can provide valuable insights into the typical price ranges and the variation in prices observed in Germany within the dataset.
# Distribution of Prices in Germany

plt.figure(figsize=(12, 6))
sns.histplot(data_df['Price.DE.'].dropna(), bins=20, kde=True, color='salmon')
plt.title('Distribution of Prices in Germany')
plt.xlabel('Price in Germany')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Fast Charging Time which is assessing the typical range of charging durations

plt.figure(figsize=(10, 6))
sns.histplot(data_df['Fast_charge'].dropna(), bins=20, kde=True, color='lightgreen')
plt.title('Distribution of Fast Charging Time')
plt.xlabel('Fast Charging Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of Range vs Efficiency

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Range', y='Efficiency', data=data_df, hue='Battery', palette='viridis', size='Top_speed', sizes=(20, 200))
plt.title('Scatter Plot of Range vs Efficiency')
plt.xlabel('Range (km)')
plt.ylabel('Efficiency (Wh/km)')
plt.show()
```

#### As per plot we can analyse there few cars with top speed of 320 kmph with range of 380-420 km having 100 wh/km, that means car enthusiast will like to ride this category of cars.

```python
# Distribution of Battery capacity

plt.figure(figsize=(10, 6))
sns.histplot(data_df['Battery'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Battery Capacity')
plt.xlabel('Battery Capacity (kWh)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of the count of cars by Top Speed category

plt.figure(figsize=(12, 6))
top_speed_counts = data_df['Top_speed'].value_counts()
sns.barplot(x=top_speed_counts.index, y=top_speed_counts.values, palette='pastel')
plt.title('Count of Cars by Top Speed Category')
plt.xlabel('Top Speed (km/h)')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot with hue based on Top Speed category

selected_columns = ['Battery', 'Efficiency', 'Range', 'Top_speed', 'acceleration..0.100.']
sns.pairplot(data_df[selected_columns], hue='Top_speed', palette='pastel')
plt.suptitle('Pairplot with Hue based on Top Speed Category', y=1.02)
plt.show()
```

```python
# Bar plot of the count of cars by Battery category

plt.figure(figsize=(12, 6))
battery_category_counts = data_df['Battery'].value_counts()
sns.barplot(x=battery_category_counts.index, y=battery_category_counts.values, palette='pastel')
plt.title('Count of Cars by Battery Category')
plt.xlabel('Battery Category')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart of the distribution of Car Manufacturers

car_manufacturer_counts = data_df['Car_name'].apply(lambda x: x.split(' ')[0]).value_counts()
plt.figure(figsize=(10, 10))
plt.pie(car_manufacturer_counts, labels=car_manufacturer_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Car Manufacturers')
plt.show()
```

```python
# Boxplot of Acceleration Time by Top Speed

plt.figure(figsize=(12, 6))
sns.boxplot(x='Top_speed', y='acceleration..0.100.', data=data_df, palette='pastel')
plt.title('Boxplot of Acceleration Time by Top Speed')
plt.xlabel('Top Speed (km/h)')
plt.ylabel('Acceleration Time (0-100 km/h)')
plt.show()
```

```python
# Jointplot of Range and Efficiency into this plot help in understanding how data points distributed
# and density of data points concerning the Range-Efficiency.

sns.jointplot(x='Range', y='Efficiency', data=data_df, kind='hex', color='skyblue')
plt.suptitle('Jointplot of Range and Efficiency', y=1.02)
plt.show()
```

```python
# KDE plot of Battery vs. Top Speed

plt.figure(figsize=(10, 6))
sns.kdeplot(x='Battery', y='Top_speed', data=data_df, cmap='Blues', fill=True)
plt.title('Kernel Density Estimation (KDE) Plot of Battery vs. Top Speed')
plt.xlabel('Battery Capacity (kWh)')
plt.ylabel('Top Speed (km/h)')
plt.show()
```

```python
# Extracting the car manufacturer from Car_name

data_df['Car_Manufacturer'] = data_df['Car_name'].apply(lambda x: x.split(' ')[0])

# Bar plot of the top 10 car manufacturers

top_manufacturers = data_df['Car_Manufacturer'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_manufacturers.index, y=top_manufacturers.values, palette='viridis')
plt.title('Top 10 Car Manufacturers')
plt.xlabel('Car Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(18, 8))
sns.boxplot(x='Car_Manufacturer', y='Range', data=data_df, palette='pastel')
plt.title('Range Distribution by Car Manufacturer')
plt.xlabel('Car Manufacturer')
plt.ylabel('Range (km)')
plt.xticks(rotation=45, ha='right')
plt.show()
```
